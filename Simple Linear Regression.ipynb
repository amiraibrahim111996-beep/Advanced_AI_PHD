```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset=pd.read_csv(r'D:\PHD\AI\dataset\Data\Salary_Data.csv')
dataset
```

```python
# when i run on this data NAN Values appeared so i used dropna
```

```python
dataset = dataset.dropna()
```

```python
X = dataset.iloc[:,[4]].values   # Features
y = dataset.iloc[:, [5]].values    # Target
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
slope = model.coef_[0]
intercept = model.intercept_
```

```python
print("slope",slope)
print("intercept",intercept)
```

```python
employee_salery=model.predict([[11]])
```
